```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import datetime

```

# IMPORTING DATASET



```python
Product_sales= pd.read_excel("SCM Project.xlsx")
Product_sales
```

```python
Product_sales.head(25)
```

```python
# check our datatypes

Product_sales.dtypes
```

# check missing values for each column 

```python
Product_sales.isnull().sum().sort_values(ascending=False)
```

```python
## Adding new column Total_Net_AMT

Product_sales['Total_Net_Amt'] = Product_sales['QUANTITY'] * Product_sales['RATE']
Product_sales
```

```python
Product_sales.head()
```

# EXPLORATORY DATA ANALYSIS

```python
Product_sales.info()
```

```python
Product_sales.describe()
```

```python
# change columns type - Float to Int type 

Product_sales['Total_Net_Amt'] = Product_sales['Total_Net_Amt'].astype('int64')
```

```python
#Inserting new column no of sales using insert method

Product_sales.insert(7, "No of Sales", Product_sales["Total_Net_Amt"])
Product_sales.head()
```

```python
Product_sales['No of Sales'] = Product_sales['No of Sales'].astype('int64')
```

```python
# Consolidating the number of Sales Made per customer for each State

sales_per_cust = Product_sales.groupby(by=['Customer_Id','State'], as_index=False)['INVOICE_ID'].count().sort_values(by='INVOICE_ID', ascending=False)
sales_per_cust.columns = ['CustomerID', 'State', "No of Sales"]
sales_per_cust.head(10)
```

```python
# Statewise total no of count of products

Product_Count = Product_sales["State"].value_counts().nlargest(5)
print("Top 5 States Wise Count \n")
print(Product_Count)
```

```python
#visualization of total net amount of product state wise

Totalnetamount = Product_sales.groupby(['State'])['Total_Net_Amt'].sum().nlargest(5)
print("Total Net Amount for Top 5 States\n")
print(Totalnetamount)
plt.figure(figsize=(15,8))
sns.set(style="darkgrid")
GraphData=Product_sales.groupby(['State'])['Total_Net_Amt'].sum().nlargest(5)



GraphData.plot(kind='bar', color=['orange', 'red', 'green', 'blue', 'cyan'])
plt.xlabel("State ")
plt.ylabel("Total_Net_Amt ")

plt.show()




```

```python
#visualization of total number of sales state wise

NoOfSales = Product_sales.groupby(['State'])['No of Sales'].sum().nlargest(5)
print("Total NoOfSales for Top 5 States\n")
print(NoOfSales)
plt.figure(figsize=(15,8))
sns.set(style="darkgrid")
GraphData=Product_sales.groupby(['State'])['No of Sales'].sum().nlargest(5)
GraphData.plot(kind='bar', color=['orange', 'green', 'purple', 'blue', 'cyan'])

plt.ylabel('NoOfSales')
plt.xlabel('State Name')
```

```python
#visualization of total number of Quantity count state wise

Quantities = Product_sales["State"].value_counts().nlargest(5)
print("Top 5 States Wise customer Count \n")
print(Quantities)
plt.figure(figsize=(15,8))
sns.set(style="darkgrid")
GraphData=Product_sales.groupby(['State'])['QUANTITY'].sum().nlargest(5)
GraphData.plot(kind='bar', color=['orange', 'cyan', 'blue', 'green', 'purple'])

plt.ylabel('No of quantities')
plt.xlabel('State Name')


```
